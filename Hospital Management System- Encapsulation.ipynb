```python
class patient:
    def __init__(self,name,age):
        self.name= name
        self.age= age
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def set_name(self,name):
        self.name= name
    def set_age(self,age):
        self.age=age
    def display(self):
        print("patient name:", self.name)
        print("age:",self.age)
        
class doctor:
    def __init__(self, name, specialization):
        self._name = name
        self._specialization = specialization
    def display(self):
        print("doctor Name:", self._name)
        print("Specialization:", self._specialization)

p1 = patient("mr.george", 67)
d1 = doctor("dr.joe", "Cardiology")

print("patient details:")
p1.display()
p1.set_name("mr.bush")
p1.set_age(74)
print("updated patient details:")
print("Name:", p1.get_name())
print("Age:", p1.get_age())
print("doctor details:")
d1.display()
```
