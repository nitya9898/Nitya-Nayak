```python
#It is a Python program that uses a user-defined function to display personalized greetings and allows multiple users using a loop
# Function to greet user
def greet(name, city):
    print("Hello, " + name)
    
    if city != "":
        print("Nice to meet you from " + city)
    
    print("Welcome to the Greeting App, Hope you have a good day")
    
# Main program
def main():
    print("Greeting App")
    
    while True:
        name = input("Enter your name: ")
        city = input("Enter your city (optional): ")
        
        greet(name, city)
        
        choice = input("Do you want to greet another user? (yes/no): ").lower()
        
        if choice != "yes":
            print("Thank you for using the Greeting App!")
            break


# Run the program
main()


#This project is a simple function-based program that greets users.

#It uses a function greet(name, city) to display a personalized message.
#The program takes user input (name and optional city).
#If the city is entered, it shows an extra message.
#A while loop allows greeting multiple users.
#The program stops when the user chooses “no”.
```
