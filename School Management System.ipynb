```python
class person:
    def __init__(self, name):
        self.name = name
    def show_name(self):
        print('name:', self.name)
        
class staff(person):
    def s(self, staffid):
        self.staffid = staffid
        
class teacher(staff):
    def t(self, name, staffid, subject):
        self.name = name
        self.staffid = staffid
        self.subject = subject 
    def show_teacher(self):
        self.show_name()
        print('staffid:', self.staffid)
        print('subject:', self.subject)
        
class principal(staff):
    def p(self, name, staffid, school):
        self.name = name
        self.staffid = staffid
        self.school = school       
    def show_principal(self):
        self.show_name()
        print('staffid:', self.staffid)
        print('school:', self.school)
        
class student(person):
    def s(self, name, studentid):
        self.name = name
        self.studentid = studentid
    def show_student(self):
        self.show_name()
        print('studentid:', self.studentid)
        
t1 = teacher("temp")
p1 = principal("temp")
s1 = student("temp")
t1.t('mr.doshi', 't11', 'chemistry')
p1.p('dr.shah', 'p01', 'XYZ school')
s1.s('Nitya Nayak', '101')
t1.show_teacher()
p1.show_principal()
s1.show_student()
```
