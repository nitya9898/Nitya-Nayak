```python
#Visitor Management Sytem
## This program records visitor details, checks the purpose of visit, and gives instructions accordingly.


print ("VISITOR MANAGEMENT SYSTEM")

visitorname = input('enter visitor name')
hostname = input('enter hostname:')
purpose = input('enter the purpose of visit(meeting/delivery/interview)')
In_time = input('Enter the time of visit (HH:MM): ')
Out_time = input('Enter the exit time (HH:MM): ')

if purpose =='meeting':
    print('Entry is allowed, please proceed to the meeting room')
elif purpose =='delivery':
    print('Please leave the package at the reception desk')
elif purpose =='interview':
    print('Please wait, HR will call you shortly')
else:
    print('Entry is prohibited')


print("---Visitor Details---")
print('visitorname:', visitorname)
print('hostname:', hostname)
print('purpose:', purpose)
print('In Time:', In_time)
print('Out Time:', Out_time)

print("\nThank you for visiting! Hope your visit goes smoothly.")
```
