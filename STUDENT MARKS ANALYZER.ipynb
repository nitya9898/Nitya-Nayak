Student marks analyzer project

```python
# This project and the below mentioned functions finds the lowest, highest, total, and average marks from a list of scores.

def get_lowest(scores):
    # Start with the first score as the lowest
    low = scores[0]  
    for value in scores:
        if value < low:
            low = value
    print("Lowest score is:", low)
def get_highest(scores):
    # Start with the first score as the highest
    high = scores[0] 
    for value in scores:
        if value > high:
            high = value
    print("Highest score is:", high)
def calculate_total(scores):
    # Add up all the scores
    total_marks = 0
    for value in scores:
        total_marks += value
    print("Total marks is:", total_marks)
    return total_marks   # return so average can use it
def calculate_average(scores):
    # Use the total function to avoid repeating code
    total_marks = calculate_total(scores)
    avg = total_marks / len(scores)
    print("Average marks is:", avg)
    
# Example dataset (marks of students)
scores = [2, 458, 78, 2359, 4, 8424]

# Function calls
get_lowest(scores)
get_highest(scores)
calculate_total(scores)
calculate_average(scores)

#Function explanation
# get_lowest(scores)- Starts with the first score, checks each value, and updates whenever it finds a smaller one. Prints the lowest score.
# get_highest(scores)- Starts with the first score, compares each value, and updates whenever it finds a larger one. Prints the highest score.
# calculate_total(scores)- Adds all the scores together in a loop. Prints the total and also returns it so other functions (like average) can use it.
#calculate_average(scores)-Calls calculate_total(scores) to get the sum, divides by the number of scores, and prints the average.
```
